### Q1 (b-2)

```python
# take t=2 as the spinned up state
initState = truthState[spinUpTimeStep:, :][0]
plt.plot(initState)
plt.show()
```

```python
numEnsemble = 50
ensemMemResults = []
for i in range(numEnsemble):
    print(f"========== Mem {i} ==========")
    initPerturb = initState + np.random.normal(mean, std, size=nGrid)
    stateGenerator = DataGenerator(nTimeStep=nTimeStep, dt=dt)
    stateGenerator.setupLorenz96(initPerturb, force=force)
    ensemState = stateGenerator.getSeriesTruth()
    #np.save(f"./dat_F8/lorenz96-{i:06d}.npy", ensemState)
    ensemMemResults.append(ensemState)
```

```python
def rmse(truth, mem):
    return (np.mean((truth-mem)**2))**(1/2)
```

```python
truthState = truthState[spinUpTimeStep:, :]
```

```python
rmseRecord = np.zeros(shape=(len(time), numEnsemble))
for i in range(nTimeStep):
    for memIdx, mem in enumerate(ensemMemResults):
        rmseRecord[i, memIdx] = rmse(truthState[i], mem[i])
```

```python
meanError = rmseRecord.mean(axis=1)
plt.figure(figsize=(15, 8))
plt.grid(True)
plt.plot(time, np.log(meanError**2))
#plt.yscale("log")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Time", fontsize=15)
plt.ylabel(r"Log($E^2$)", fontsize=15)
plt.show()
```

```python
doubleError = 10**(np.log10(rmseRecord.mean(axis=1)[0]) + np.log10(2))
print(time[np.argmin(np.abs(meanError-doubleError))])
```
